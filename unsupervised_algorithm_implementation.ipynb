### **Data Cleaning**

datacleaning.jpg

The first preprocessing step is data cleaning. It involves handling missing values, removing outliers, and addressing any inconsistencies or errors in the data. üë©‚Äçüíª

**Missing Values:** Missing data can be problematic for analysis and modeling. One common approach is to impute missing values by replacing them with estimated values based on the available data. This can be done using techniques like mean imputation (replacing missing values with the mean of the feature) or regression imputation (predicting missing values based on other variables). Alternatively, if the missing values are limited in number, you may choose to remove the corresponding rows or columns from the dataset. üìä

**Outlier Removal:** Outliers are extreme values that deviate significantly from the normal pattern of the data. They can negatively impact the analysis and modeling process. Outliers can be identified using statistical methods such as the Z-score or the interquartile range (IQR). Once identified, outliers can be removed or transformed to minimize their influence on the analysis. üìâ

Data Consistency: Inconsistencies or errors in the data, such as incorrect or inconsistent formatting, inconsistent encoding, or duplicate records, should be addressed. These issues can be resolved by performing data validation checks, standardizing the data format, and resolving any inconsistencies or duplicates. üîç

Data cleaning is a crucial step in preparing the data for further analysis and modeling. By handling missing values, removing outliers, and ensuring data consistency, you can improve the quality and reliability of your data, leading to more accurate and meaningful results. üßπ

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.impute import SimpleImputer
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load data using Scikit-Learn (Wine dataset)
wine = load_wine()
X, y = wine.data, wine.target

# Step 2: Convert data to a pandas DataFrame
df = pd.DataFrame(data=X, columns=wine.feature_names)

# Step 3: Create missing values in the DataFrame
# Here, we set 10% of the data in each feature to NaN (missing value)
np.random.seed(42)
mask = np.random.rand(*df.shape) < 0.1
df_with_missing = df.copy()
df_with_missing[mask] = np.nan

# Step 4: Preprocess missing values using mean imputation
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df_with_missing), columns=df.columns)

# Step 5: Outlier removal using Z-score method
z_scores = np.abs(stats.zscore(df_imputed))
outlier_mask = (z_scores < 3).all(axis=1)  # Keep samples with all features within 3 standard deviations
df_no_outliers = df_imputed[outlier_mask]

# Step 6: Visualization of different techniques
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original DataFrame with missing values
sns.heatmap(df_with_missing.isnull(), cbar=False, cmap='viridis', ax=axes[0, 0])
axes[0, 0].set_title("Original DataFrame\nwith Missing Values")

# DataFrame after Mean Imputation
sns.heatmap(df_imputed.isnull(), cbar=False, cmap='viridis', ax=axes[0, 1])
axes[0, 1].set_title("DataFrame after\nMean Imputation")

# Original DataFrame (without outliers)
sns.boxplot(data=df, ax=axes[1, 0])
axes[1, 0].set_title("Original DataFrame\n(Box Plot)")

# DataFrame after Outlier Removal
sns.boxplot(data=df_no_outliers, ax=axes[1, 1])
axes[1, 1].set_title("DataFrame after\nOutlier Removal")

plt.tight_layout()
plt.show()
```

**Explanation:**

Original DataFrame with Missing Values (Heatmap):

1. In the first heatmap, we visualize the original DataFrame with missing values. NaN values are represented in dark colors (purple in this case).

 * Each row corresponds to a sample (wine) from the dataset, and each column represents a feature (chemical analysis) of the wine.
 * The dark cells indicate the missing values, where we deliberately introduced NaNs in approximately 10% of the data.
 * This visualization helps us identify the locations of missing values in the dataset, which is essential for preprocessing.

2. DataFrame after Mean Imputation (Heatmap):

 * In the second heatmap, we visualize the DataFrame after applying mean imputation to handle the missing values.
 * After mean imputation, the NaN values are replaced with the mean value of their respective features.
 * As a result, all the dark cells (NaN values) in the original heatmap are now filled with mean values, indicated by different colors.
 * This visualization shows how mean imputation replaces missing values, ensuring a complete dataset for unsupervised learning.

3. Original DataFrame (Box Plot):

 * In the first box plot, we visualize the original DataFrame without any preprocessing, including missing values and outliers.
 * Each box plot represents the distribution of a feature across all wine samples.
 * Outliers are represented as individual points outside the whiskers of the box plot.
 * The box shows the interquartile range (IQR), and the whiskers extend to the data within 1.5 times the IQR.
 * This visualization helps us identify outliers in each feature of the dataset.

4. DataFrame after Outlier Removal (Box Plot):

 * In the second box plot, we visualize the DataFrame after applying the Z-score method for outlier removal.
 * The Z-score method identifies outliers based on how many standard deviations they deviate from the mean.
 * Samples with any feature outside 3 standard deviations are considered outliers and removed from the dataset.
 * As a result, the box plots show a reduced number of outliers compared to the original box plots.
 * This visualization demonstrates how the Z-score method removes outliers and improves the data quality for unsupervised learning.

### **Data Scaling/ Normalization**


Data scaling/normalization involves transforming the numerical features of the dataset to a common scale. This step ensures that all features have a similar range, preventing any one feature from dominating the analysis due to its larger scale. Scaling is particularly important for algorithms that rely on distance calculations or optimization techniques. ‚ú®üìä

Two common methods of scaling/normalization are:

Min-Max Scaling: This method rescales the data to a specific range, typically between 0 and 1. It involves subtracting the minimum value of the feature and dividing it by the difference between the maximum and minimum values. üìâüî¢

Standardization: This method transforms the data to have a mean of 0 and a standard deviation of 1. It involves subtracting the mean of the feature and dividing it by the standard deviation. üìèüî¢

The choice between these methods depends on the specific characteristics of the data and the requirements of the problem. ü§îüí°

It's important to note that the preprocessing steps may vary depending on the specific analysis or modeling techniques being employed, and it's always recommended to tailor the preprocessing steps to the specific needs of the data and problem at hand. üõ†Ô∏èüîé

### **Min-Max Scaling**





Min-max scaling, also known as normalization, is a common data preprocessing technique used to transform numerical data to a specific range, typically between 0 and 1. üìäüî¢





**The process involves the following steps:**

1. Identify the minimum and maximum values of the feature you want to scale. üìàüìâ

2. Apply the min-max scaling formula to each value in the feature:

 min_max_scaled_value = (value - min_value) / (max_value - min_value)

3. This formula subtracts the minimum value from each data point and divides it by the range (the difference between the maximum and minimum values). It scales the values proportionally within the range of 0 to 1. üî¢üîç

4. Repeat the above steps for all the features you want to scale. üîÑ

**Min-max scaling has several benefits:**

1. It brings different features onto a comparable scale, ensuring that no single
feature dominates the analysis or modeling process due to its larger value range. ‚öñÔ∏èüîÄ

2. It preserves the original distribution and relationships between the data points while transforming them to a common range. üìäüîÑ

3. It can help improve the performance of certain machine learning algorithms, especially those that rely on distance calculations or optimization techniques. ü§ñüî¨

Using min-max scaling can be beneficial in various scenarios, but it's important to consider the specific characteristics of the data and the requirements of the problem at hand. ü§îüí°

### **Standardization**

Standardization, also known as z-score normalization, is a preprocessing technique used to transform the features of a dataset to have a mean of 0 and a standard deviation of 1. It is a common practice in machine learning and data analysis to bring variables onto a common scale for fair comparison and to avoid certain features dominating the analysis due to their larger scales.


The process of standardization involves the following steps:

1. Calculate the mean (average) of the feature.

2. Calculate the standard deviation, which measures the spread or dispersion of the feature's values.

3. For each data point in the feature, subtract the mean and divide by the standard deviation to obtain the standardized value.

By standardizing the features, the data is transformed to have a standard distribution, with values centered around 0 and a spread determined by the standard deviation. This allows for easier interpretation and comparison of the data, and it can be beneficial for certain machine learning algorithms that assume standardized input.

Standardization is particularly useful when features have different scales or units of measurement, as it brings them to a comparable level. It also helps in handling outliers by reducing their impact on the data analysis.

Let's dive into standardization with an example.

Suppose you have a dataset containing information about students' test scores and their ages. Here's a sample of the data:

tab4.png

In standardization, we transform the data in such a way that it has a mean of 0 and a standard deviation of 1. This is also referred to as z-score normalization.

To standardize a feature, we follow these steps:

1. Calculate the mean of the feature.

2. Calculate the standard deviation of the feature.

3. Apply the formula to calculate the standardized values for each data point.

Let's standardize the Test Score feature:

1. Calculate the mean of the Test Score:
  
   mean = (80 + 90 + 70 + 85 + 95) / 5 = 84

2. Calculate the standard deviation of the Test Score:

   standard_deviation = sqrt(((80 - 84)^2 + (90 - 84)^2 + (70 - 84)^2 + (85 - 84)^2 + (95 - 84)^2) / 5) ‚âà 8.37

3. Apply the formula to calculate the standardized values for each test score:
   
   standardized_test_score = (value - mean) / standard_deviation

   For example, for the first student with a test score of 80:

   standardized_test_score = (80 - 84) / 8.37 ‚âà -0.48




By performing the same calculation for all test scores in the dataset, we obtain the standardized Test Score column:

It's important to note that standardization does not change the underlying distribution of the data, but rather re-scales it to a standardized range.

## **Evaluating Clustering and Dimensionality Reduction Models**

When evaluating clustering and dimensionality reduction models, several commonly used metrics and techniques can provide insights into the quality and performance of the models. Three such evaluation methods are the silhouette score, the elbow method, and reconstruction error.

 Let's explore each of these techniques: üòäüìä

### **Silhouette Score**

The silhouette score is a metric used to evaluate the quality of clustering results. It measures how well each sample in a cluster fits with its own cluster compared to other clusters. A higher silhouette score indicates better-defined and well-separated clusters.

The silhouette score ranges from -1 to 1:

1. A score close to 1 indicates that the sample is well-matched to its own
  cluster and poorly matched to neighboring clusters.

2. A score close to 0 indicates that the sample is on or very close to the
   decision boundary between two neighboring clusters.

3. A score close to -1 indicates that the sample is likely assigned to the  
   wrong cluster.

Here's an example of calculating the silhouette score using scikit-learn:

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data with 2 clusters
X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=42)

# Apply clustering algorithm
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Obtain cluster assignments
labels = kmeans.labels_

# Calculate the silhouette score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)
```

In this example, we first generate synthetic data using the make_blobs function from scikit-learn, creating two clusters. We then apply the K-Means clustering algorithm to cluster the data points. The labels variable contains the cluster assignments obtained from K-Means.

Next, we calculate the silhouette score using the silhouette_score function. The function takes the feature matrix X and the cluster labels labels as inputs. It computes the average silhouette score across all samples, representing the overall quality of the clustering.

Finally, we print the silhouette score to assess the clustering results. A higher score closer to 1 indicates better-defined clusters.

The silhouette score is a useful metric to compare different clustering algorithms, evaluate the number of clusters, and assess the quality of clustering results. However, it should be used in conjunction with other evaluation methods and domain knowledge for a comprehensive analysis of the clustering performance. ‚ú®üîç






### **Elbow Method**

The elbow method is a technique used to determine the optimal number of clusters in a clustering algorithm. It helps identify the "elbow" point on a plot of the within-cluster sum of squares (WCSS) or inertia against the number of clusters. The elbow point represents a balance between the model's ability to minimize within-cluster variance and avoid overfitting.

The steps involved in the elbow method are as follows:

1. Apply a clustering algorithm (e.g., K-Means) to the dataset for a range of
 different cluster numbers (k values).

2. For each k value, calculate the WCSS or inertia, which represents the sum of
 squared distances between each data point and its centroid within the cluster.

3. Plot the WCSS or inertia values against the corresponding k values.

4. Analyze the resulting plot and identify the elbow point, where the rate of
 improvement in WCSS or inertia decreases significantly. This point indicates a trade-off between the number of clusters and the overall model performance.

Here's an example of using the elbow method with the K-Means algorithm in scikit-learn:

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=100, n_features=2, centers=5, random_state=42)

# Define range of k values
k_values = range(1, 10)
inertias = []

# Apply K-Means for different k values
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method')
plt.show()
```

In this example, we first generate synthetic data using the make_blobs function from scikit-learn. We then define a range of k values (number of clusters) to evaluate.

Next, we iterate over the k values and apply the K-Means algorithm for each k value. We calculate and store the corresponding WCSS (inertia) using the inertia_ attribute of the K-Means model.

Finally, we plot the elbow curve, showing the WCSS values against the number of clusters. The elbow point on the curve indicates the optimal number of clusters.

By observing the plot, you can identify the elbow point where the WCSS improvement begins to level off. This suggests the appropriate number of clusters to use for the dataset. However, it's important to note that the elbow method is subjective, and domain knowledge should be considered in conjunction with this technique to make informed decisions about the optimal number of clusters. üîç‚õ∞Ô∏èüí°

### **Reconstruction Error**

Reconstruction error is a metric used to evaluate the quality of dimensionality reduction techniques, particularly those that involve data reconstruction. It measures the discrepancy or error between the original data and the reconstructed data after reducing the dimensions. üîçüèóÔ∏èüìâ

The reconstruction error quantifies how well the reduced-dimensional representation captures the important information from the original data. A lower reconstruction error indicates a more accurate representation of the original data and implies that less information was lost during the dimensionality reduction process. üìäüî¢üîç

One common dimensionality reduction technique that uses reconstruction is Principal Component Analysis (PCA). PCA aims to find orthogonal directions (principal components) that capture the maximum variance in the data. The reconstruction process involves projecting the reduced-dimensional data back into the original high-dimensional space. üí°üìâüîÅ

To calculate the reconstruction error, you need to compare the original data with the reconstructed data and measure their similarity. Mean squared error (MSE) is often used as the reconstruction error metric.

Here's an example of calculating the reconstruction error using PCA in scikit-learn:

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.metrics import mean_squared_error

# Load the Iris dataset
data = load_iris()
X = data.data

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Reconstruct the data
X_reconstructed = pca.inverse_transform(X_reduced)

# Calculate the reconstruction error (MSE)
reconstruction_error = mean_squared_error(X, X_reconstructed)
print("Reconstruction Error:", reconstruction_error)
```

In this example, we load the Iris dataset using load_iris from scikit-learn. We then apply PCA to reduce the dimensionality of the data to two components.

After dimensionality reduction, we reconstruct the data by applying the inverse transformation (inverse_transform) to the reduced data, obtaining X_reconstructed.

Finally, we calculate the reconstruction error using the mean squared error (mean_squared_error) between the original data (X) and the reconstructed data (X_reconstructed).

A lower reconstruction error indicates a more accurate ‚úÖ representation of the original data. The reconstruction error can be used to assess the quality of dimensionality reduction and compare different techniques or configurations. However, it is important to consider other factors such as the retained variance and the specific objectives of the dimensionality reduction task.üí°

## **Deploying unsupervised models in production**

Deploying unsupervised models in production involves making the models accessible üëå and operational in a production environment, allowing them to generate predictions üí° or perform tasks on new, üîç unseen data.

 There are two common approaches for deploying unsupervised models: batch processing and real-time streaming. Let's explore each of these approaches:

### **Batch Proccessing**

Batch processing is a technique where üî¢ data is collected and processed in groups or batches  rather than in real-time ‚è≤ or as it arrives. In this approach, data is accumulated over a period of time, and the processing is performed on the entire batch at once. Batch processing is commonly used when immediate results or real-time analysis are not necessary, and it offers several advantages such as ‚úÖ efficient resource utilization and the ability to handle üí≠ large volumes of data.

Here's a simplified example of batch processing:

Let's say you have a dataset containing customer transaction records for an e-commerce platform. The goal is to perform customer segmentation using unsupervised clustering algorithms. Here's how batch processing can be applied:



1. **Data Collection:** Accumulate the transaction data over a specific time
 period, such as a day or a week.

2. **Data Preprocessing:** Perform any necessary data cleaning, normalization, or feature engineering steps on the collected data. This may involve removing missing values, scaling the features, or encoding categorical variables.

3. **Clustering Model Training:** Apply the unsupervised clustering algorithm, such as K-Means, to the preprocessed data. Train the model using the entire batch of data.

4. **Clustering Model Evaluation:** Assess the quality of the clustering results using appropriate evaluation metrics, such as silhouette score or within-cluster sum of squares (WCSS).

5. **Post-Processing and Analysis:** Analyze the clustering results to gain insights and understanding of customer segments. This may involve visualizing the clusters, computing cluster centroids, or extracting meaningful patterns from the data.

6. **Storage or Output:** Store the clustering results or use them for downstream tasks, such as targeted marketing campaigns, personalized recommendations, or customer behavior analysis.

In this example, the batch processing approach allows you to process a large volume of transaction data collected over a specific time period. The entire dataset is processed as a batch, and the clustering model is trained and applied to the complete dataset at once. This enables efficient resource utilization as the computations can be performed in bulk, and it eliminates the need for real-time or continuous data processing.

Batch processing is commonly used in scenarios where real-time decision-making is not required, and it offers benefits such as the ability to process large volumes of data efficiently, simplified resource management, and the opportunity for offline analysis and optimization.

## **Best practices for working with unsupervised data**

When working with unsupervised data, there are several best practices to consider. Here are three important ones: outlier detection, data scaling, and handling imbalanced datasets.

### **Outlier Detection**

outlier.jpg


Outliers are data points that deviate significantly from the majority of the data. They can have a significant impact on the performance and results of unsupervised algorithms. It's important to identify and handle outliers appropriately. Some best practices for outlier detection include:

1. Visual exploration: Plotting the data and visually inspecting for any data points that appear significantly different from the majority.

2. Statistical methods: Utilizing statistical techniques such as z-scores, percentiles, or box plots to detect outliers based on the distribution of the data.

3. Unsupervised outlier detection algorithms: Applying algorithms like DBSCAN (Density-Based Spatial Clustering of Applications with Noise) or Isolation Forest to automatically identify outliers in the data.

Let's consider an example where we have a dataset of random numbers, and we want to detect and visualize outliers using the z-score method:



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate random data
np.random.seed(42)
data = np.random.normal(0, 1, 1000)

# Calculate z-scores
z_scores = zscore(data)

# Set a threshold for outlier detection
threshold = 3

# Identify outliers
outliers = np.abs(z_scores) > threshold

# Visualize the data and outliers
plt.figure(figsize=(10, 6))
plt.scatter(range(len(data)), data, c=outliers, cmap='coolwarm')
plt.xlabel('Index')
plt.ylabel('Data')
plt.title('Outlier Detection')
plt.colorbar(label='Outlier')
plt.show()

```

In this code, we generate a dataset of 1000 random numbers using np.random.normal. We then calculate the z-scores of the data using the zscore function from the scipy.stats module.

Next, we set a threshold for outlier detection. Typically, a z-score above a certain threshold (e.g., 3) is considered an outlier.

We identify the outliers by comparing the absolute values of the z-scores to the threshold. If the absolute z-score exceeds the threshold, the corresponding data point is marked as an outlier.

Finally, we visualize the data using a scatter plot, where outliers are colored differently using the c parameter of the scatter function. The color bar on the side represents whether a point is an outlier or not.

This code provides a basic example of outlier detection using the z-score method and visualizing the outliers in the dataset. You can explore other outlier detection techniques and adjust the threshold or visualization settings based on your specific requirements and dataset characteristics.

### **Data Scaling**

scalee.png

Unsupervised algorithms are sensitive to the scale and magnitude of the features. Data scaling is important to ensure that all features have a similar range or distribution. Some best practices for data scaling that we have already discussed before include:

1. Standardization: Scaling the data to have zero mean and unit variance using techniques like StandardScaler from scikit-learn.

2. Min-Max scaling: Scaling the data to a specific range, typically between 0 and 1, using techniques like MinMaxScaler from scikit-learn.

3. Robust scaling: Scaling the data using robust statistics that are less affected by outliers, such as the RobustScaler from scikit-learn.

Choosing the appropriate scaling technique depends on the characteristics of the data and the requirements of the algorithm being used.



**Standardization**  is a common technique used to scale features by subtracting the mean and dividing by the standard deviation. This process transforms the data to have zero mean and unit variance.

 Here's an example of how to use StandardScaler for standardization in Python:

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Create a sample dataset
data = np.array([[1.2], [2.4], [3.6], [4.8]])

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

# Print the original and scaled data
print("Original Data:")
print(data)
print("\nScaled Data:")
print(scaled_data)
```

In this example, we have a sample dataset data with four values. We import the StandardScaler from scikit-learn.

Next, we initialize the StandardScaler by creating an instance of the StandardScaler class.

We then fit the scaler to the data and transform it using the fit_transform method. This computes the mean and standard deviation from the data and applies the standardization transformation.

Finally, we print the original data and the scaled data to observe the effect of standardization. The scaled data will have a mean of zero and a standard deviation of one.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Create a sample dataset
data = np.array([[1.2], [2.4], [3.6], [4.8]])

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

# Print the original and scaled data
print("Original Data:")
print(data)
print("\nScaled Data:")
print(scaled_data)
```

### **Handling Imbalanced Datasets**

an_overview_on_imbalanced_dataset_in_ml_and_techniques_to_handling_it.jpg

Handling imbalanced datasets is an üí° important consideration in ü§ñ unsupervised learning, as it can impact the performance and ‚úÖ accuracy of the algorithms. Imbalanced datasets refer to datasets where the distribution of classes or clusters is uneven, with one or more classes having significantly fewer samples than others. This can lead to biased results and üòµ difficulties in detecting the minority class or cluster.

Here are some techniques for handling imbalanced datasets:

**Oversampling:**

1. Oversampling involves increasing the number of instances in the minority class to balance the dataset.

2. This can be done by duplicating existing instances or generating synthetic samples using techniques like SMOTE (Synthetic Minority Over-sampling Technique).

3. Oversampling helps the algorithm to learn from the minority class more effectively.

0_FeIp1t4uEcW5LmSM.png

**Undersampling:**

1. Undersampling involves reducing the number of instances in the majority class to balance the dataset.

2. Randomly or strategically removing instances from the majority class can be used to achieve a balanced distribution.

3. Undersampling reduces the dominance of the majority class and ensures the algorithm is exposed to the minority class adequately.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

# Step 1: Generate imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=2, weights=[0.1, 0.9], n_informative=3, n_redundant=1,
                           flip_y=0, n_features=5, n_clusters_per_class=1, n_samples=1000, random_state=42)

# Function to plot scatter plot
def plot_scatter(X, y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.5)
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.5)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

# Function to print sampling output
def print_sampling_output(X_resampled, y_resampled, technique):
    print(f"Sampling Technique: {technique}")
    print(f"Original Class Distribution: {dict(zip(*np.unique(y, return_counts=True)))}")
    print(f"Resampled Class Distribution: {dict(zip(*np.unique(y_resampled, return_counts=True)))}")
    print("=" * 40)

# Step 2: Plot original dataset using scatter plot
plot_scatter(X, y, 'Original Dataset - Scatter Plot')

# Step 3: Perform Random Oversampling
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(X, y)
plot_scatter(X_ros, y_ros, 'Random Oversampling - Scatter Plot')
print_sampling_output(X_ros, y_ros, 'Random Oversampling')

# Step 4: Perform Random Undersampling
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X, y)
plot_scatter(X_rus, y_rus, 'Random Undersampling - Scatter Plot')
print_sampling_output(X_rus, y_rus, 'Random Undersampling')

# Step 5: Perform SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
plot_scatter(X_smote, y_smote, 'SMOTE - Scatter Plot')
print_sampling_output(X_smote, y_smote, 'SMOTE')
```

**Ensemble Methods:**

1. Ensemble methods combine multiple unsupervised learning models or techniques to handle imbalanced datasets effectively.

2. By combining the results of multiple models, ensemble methods can improve the overall performance and address class imbalance.

3. Examples of ensemble methods for unsupervised learning include clustering ensembles and anomaly detection ensembles.



**Anomaly Detection:**

1. Anomaly detection approaches can be used to treat the imbalanced class as an anomaly.

2. Techniques like One-Class SVM (Support Vector Machine) or Local Outlier Factor can identify instances that deviate significantly from the majority class.

3. Anomaly detection can help identify and handle the minority class effectively.



**Evaluation Metrics:**

1. When evaluating the performance of unsupervised algorithms on imbalanced datasets, it is crucial to choose appropriate evaluation metrics.

2. Accuracy alone may be misleading due to the imbalanced nature of the dataset.

3. Metrics like precision, recall, F1-score, or area under the ROC curve (AUC-ROC) can provide a more comprehensive assessment of model performance.

Here's a simple example to illustrate the handling of an imbalanced dataset using oversampling with the SMOTE technique:

```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification

# Generate an imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=10, weights=[0.95, 0.05], random_state=42)

# Apply SMOTE oversampling
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the class distribution after oversampling
unique, counts = np.unique(y_resampled, return_counts=True)
class_distribution = dict(zip(unique, counts))
print("Class distribution after oversampling:", class_distribution)
```

In this example, we generate an imbalanced dataset using scikit-learn's make_classification function. We set the weights to [0.95, 0.05] to create a highly imbalanced dataset with a minority class representing only 5% of the samples.

We then use the SMOTE oversampling technique from the imblearn library to balance the dataset. The fit_resample method applies SMOTE oversampling to generate synthetic samples for the minority class.

Finally, we check the class distribution after oversampling to ensure that the minority class has been adequately represented.
